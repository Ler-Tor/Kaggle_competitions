```python
labels = data[data.columns[-1]].values  #Выделим значения метки класса в переменную labels
feature_matrix = data[data.columns[:-1]].values #Признаковые описания в переменную feature_matrix
```

```python
from sklearn.model_selection import train_test_split
```

```python
train_feature_matrix, test_feature_matrix, train_labels, test_labels = train_test_split(
    feature_matrix, labels, test_size=0.2, random_state=42)
```

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score

clf = KNeighborsClassifier(n_neighbors = 5, weights='uniform', algorithm='auto', leaf_size=30, p = 2, metric='minkowski',metric_params=None, n_jobs=None)
clf.fit(train_feature_matrix, train_labels)

y_pred = clf.predict(test_feature_matrix)

accuracy_score(test_labels, y_pred)
```
